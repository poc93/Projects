```python
import sklearn.preprocessing as preproc      # Standardizing all values except 'Outcome' column to avoid scale weighting issues

scaler = preproc.StandardScaler()

df_standardized = scaler.fit_transform(df[df.columns[:-1]])


df_standardized = pd.DataFrame(df_standardized, columns = df.columns[: -1])

df_standardized.head(5)
```

```python
fig = plt.figure(figsize = (8, 8))   # Checking statistical shape is same after standardization for each variable, looks good

ax = fig.gca()

df_standardized.hist(ax = ax)

plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (10, 5))           
fig.tight_layout(pad = 3.0)

df['Pregnancies'].hist(ax = ax1, bins = 20)       # side-by-side checking statistical shape is unchanged after standardization
ax1.tick_params(labelsize = 14)
ax1.set_xlabel('Pregnancies', fontsize = 14)
ax1.set_ylabel('Count', fontsize = 14)

df_standardized['Pregnancies'].hist(ax = ax2, bins = 20)
ax2.tick_params(labelsize = 14)
ax2.set_xlabel('Standardized Pregnancies', fontsize = 14)
ax2.set_ylabel('Count', fontsize = 14)
```

```python
df_standardized = pd.concat([df_standardized, df.iloc[:, -1]], axis = 1)   

df_standardized.head(5)     # Bringing the target variable into the new dataframe of focus
```

```python
from sklearn.model_selection import train_test_split  # Splitting the data into training and testing

train_data, test_data, train_label, test_label = train_test_split(df_standardized.iloc[:, :-1], 
                                                                  df_standardized.iloc[:, -1], test_size = 0.2, 
                                                                  random_state = 42)

train_data.shape, test_data.shape, train_label.shape, test_label.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier  # Bringing in k-nearest neighbors for ML modelling

neighbors = np.arange(10, 35)
train_accuracy = np.zeros(len(neighbors))
test_accuracy = np.zeros(len(neighbors))
```

```python
for i, k in enumerate(neighbors):                 
    knn = KNeighborsClassifier(n_neighbors = k)
    
    knn.fit(train_data, train_label)
    
    train_accuracy[i] = knn.score(train_data, train_label)
    
    test_accuracy[i] = knn.score(test_data, test_label)
    
    
print(train_accuracy)
print(test_accuracy)                # Running the KNN model with different k values to determine optimum value for k
```

```python
plt.figure(figsize = (10,10))

plt.title('KNN accuracy with varying numbers of k', fontsize = 20)
plt.plot(neighbors, test_accuracy, label = 'Testing Accuracy')
plt.plot(neighbors, train_accuracy, label = 'Training Accuracy')

plt.legend(prop = {'size': 20})
plt.xlabel('Number of neighbors', fontsize = 20)
plt.ylabel('Accuracy', fontsize = 20)
plt.xticks(fontsize = 20)
plt.yticks(fontsize = 20)

plt.show()                   # Displaying which number of k produces best results, looks like k = 17
```

```python
knn = KNeighborsClassifier(n_neighbors = 17)        # Performing the KNN with the optimized k

knn.fit(train_data, train_label)

train_accuracy = knn.score(train_data, train_label)
test_accuracy = knn.score(test_data, test_label)

print(train_accuracy)
print(test_accuracy)
```

```python
from sklearn.metrics import confusion_matrix   # Computing confusion matrix

prediction = knn.predict(test_data)

confusion_matrix(test_label, prediction)
```

```python
from sklearn.metrics import plot_confusion_matrix       # Plotting confusion matrix results
from sklearn.metrics import confusion_matrix
import itertools

def plot_confusion_matrix(cm, classes, normalize = False,
                         title = 'Confusion Matrix', cmap = plt.cm.Blues):
    plt.imshow(cm, interpolation = 'nearest', cmap = cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation = 45)
    plt.yticks(tick_marks, classes)
    
    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2
    
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment = 'center',
                 color = 'white' if cm[i, j] > thresh else 'black')
    
    plt.ylabel('True Label', fontsize = 30)
    plt.xlabel('Predicted Label', fontsize = 30)
    plt.tight_layout()
    plt.xticks(fontsize = 18)
    plt.yticks(fontsize = 18)

class_names = df_standardized['Outcome'].unique()

cnf_matrix = confusion_matrix(test_label, prediction)

plt.figure(figsize = (10, 8))
plot_confusion_matrix(cnf_matrix, classes = class_names)
plt.title('Confusion Matrix', fontsize = 30)

plt.show()      

    
```

```python
from sklearn.metrics import classification_report     # Classification report, recall requires improvement for 'positive' class

report = classification_report(test_label, prediction)

print(report)
```
